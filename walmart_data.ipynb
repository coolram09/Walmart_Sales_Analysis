```python
import pandas as pd
df_sales =pd.read_csv("walmart.csv")
df_sales
```

```python
#removing duplicates
df_sales.drop_duplicates(inplace=True)
df_sales.duplicated().sum()
```

```python
df_sales.shape
```

```python
df_sales.dtypes
```

```python
df_sales["unit_price"]=df_sales["unit_price"].str.replace(
    "$", "",regex=False).str.replace(
    ",","" , regex=False).astype(float)
```

```python
df_sales.unit_price.dtype
```

```python
df_sales.describe()
```

```python
df_sales["unit_price"].isnull().sum(),df_sales["quantity"].isnull().sum()
```

```python
#dropping rows with missing values
```

```python
df_sales.dropna(inplace=True)
```

```python
df_sales.isnull().sum()
```

```python
df_sales.shape
```

```python
#creating column total
```

```python
df_sales["Total_price"]=df_sales["unit_price"]*df_sales["quantity"]
```
